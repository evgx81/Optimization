```python
import numpy as np
```

```python
# Импортируем библиотеку matplotlib.pyplot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sympy as sp
```

```python
x = sp.symbols("x")
```

```python
# Заданная функция f(x)
#f = 2*x - sp.log(x)
f = x**2 + x + sp.sin(x)
```

```python
esp = 10**(-2) # Величина эпсилон
```

```python
#x0 = 0.1
x0 = -0.4 # Задаем начальную точку
```

```python
k = 0
```

```python
max_iter = 50
```

```python
x_k = x0
```

```python
df1 = sp.diff(f, x)
df2 = sp.diff(df1, x)
```

```python
x_list = []
y_list = []
```

```python
for k in range(max_iter):
    if abs(df1.subs(x, x_k)) < esp:
        x_zv = x_k
        f_x_zv = f.subs(x, x_k)
        break
        
    x_k_1 = x_k - df1.subs(x, x_k) / df2.subs(x, x_k)
    x_k = x_k_1
    
    x_list.append(x_k)
    y_list.append(f.subs(x, x_k))
    
    k = k + 1
```

```python
x_zv = x_k
f_x_zv = f.subs(x, x_k)
```

```python
x_min = x_zv
f_min = f_x_zv
```

```python
print(f"Точка минимума: {x_min}; Значение функции в точке минимума: {f_min}")
print(f"Достигнуто при k = {k}")
```

```python
#x_arr = np.linspace(0.1, 1.2)
x_arr = np.linspace(-1, 0)
```

```python
y_arr = np.array([f.subs(x, i) for i in x_arr])
```

```python
# Построение графика функции f
plt.title("График функции f")
plt.plot(x_arr, y_arr, '-g', label = 'Y(x)')
plt.scatter(x_min, f_min, marker='*', label = f"x* = {x_min}, Y(x*) = {f_min}")
plt.grid()
plt.legend()
```
